```python
import csv

# copied from https://automatetheboringstuff.com/chapter14/
def process_csv(filename):
    exampleFile = open(filename, encoding="utf-8")
    exampleReader = csv.reader(exampleFile)
    exampleData = list(exampleReader)
    exampleFile.close()
    return exampleData
```

```python
def dump(filename, column):
    data = process_csv(filename)
    rows = data[1:]
    header = data[0]
    name_pos = header.index("name")
    col_pos = header.index(column)
    for row in rows:
        print(row[name_pos] + ": " + row[col_pos])
    
dump("hurricanes.csv", "formed")
```

```python
# STEP 1: create lists of names
data = process_csv("names.csv")
header = data[0]
rows = data[1:]

# what gender are these names stereotyically associated with?
male_names = []
female_names = []

for row in rows:
    name = row[header.index("name")]
    sex = row[header.index("sex")]
    if sex == "boy":
        male_names.append(name)
    elif sex == "girl":
        female_names.append(name)
    else:
        print("OOPS!  Unexpected value...")
        
print(male_names[:5])
print(female_names[:5])
```

```python
# STEP 2: count the hurricanes
data = process_csv("hurricanes.csv")
header = data[0]
rows = data[1:]

male_deaths = 0
female_deaths = 0

for row in rows:
    name = row[header.index("name")]
    deaths = int(row[header.index("deaths")])
    if name in male_names:
        male_deaths += deaths
    elif name in female_names:
        female_deaths += deaths
        
print("Male Hurricanes:", male_deaths)
print("Female Hurricanes:", female_deaths)
```
