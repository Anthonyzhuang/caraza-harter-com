# Binary Trees

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

    def draw_edges(self, g):
        if self.left != None:
            g.edge(self.val, self.left.val, label="L")
            self.left.draw_edges(g)

        if self.right != None:
            g.edge(self.val, self.right.val, label="R")
            self.right.draw_edges(g)

    def _repr_svg_(self):
        g = Digraph()
        self.draw_edges(g)
        return g._repr_svg_()

    def dump(self):
        if self.left != None:
            self.left.dump()
        if self.right != None:
            self.right.dump()
        print(self.val)
    
root = Node("A")
root.left = Node("B")
root.right = Node("C")
root.left.right = Node("D")
root.dump() # print all the node values
root
```

# Contains

```python
from IPython.core.display import display

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

    def draw_edges(self, g):
        g.node(self.val)

        if self.left != None:
            g.edge(self.val, self.left.val, label="L")
            self.left.draw_edges(g)
            
        if self.right != None:
            g.edge(self.val, self.right.val, label="R")
            self.right.draw_edges(g)

    def _repr_svg_(self):
        g = Digraph()
        self.draw_edges(g)
        return g._repr_svg_()
    
    def __contains__(self, x):
        #display(self)
        if self.val == x:
            return True
        
        if self.left != None:
            if self.left.__contains__(x):
                return True

        if self.right != None:
            if self.right.__contains__(x):
                return True

        return False

root = Node("B")
root.left = Node("A1")
root.right = Node("C1")
root.right.left = Node("C2")
root.right.left.right = Node("C3")
root.right.left.left = Node("C4")
root.right.left.left.left = Node("C5")
root.left.right = Node("A2")
"A3" in root
```

# Practice

```python
from graphviz import Graph, Digraph

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

    def draw_edges(self, g):
        if self.left != None:
            g.edge(self.val, self.left.val, label="L")
            self.left.draw_edges(g)

        if self.right != None:
            g.edge(self.val, self.right.val, label="R")
            self.right.draw_edges(g)

    def _repr_svg_(self):
        g = Digraph()
        self.draw_edges(g)
        return g._repr_svg_()

    def count_nodes(self):
        count = 1
        if self.left != None:
            count += self.left.count_nodes()
        if self.right != None:
            count += self.right.count_nodes()
        return count

root = Node("A")
root.left = Node("B")
root.right = Node("C")
root.left.right = Node("D")
root.count_nodes()
```
